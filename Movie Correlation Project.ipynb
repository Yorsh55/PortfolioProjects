```python
# Import the packages we will use

import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import re

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None
```

```python
# Read and take a look at the data
df = pd.read_csv('/Users/jorge/Downloads/movies.csv')

df.head()
```

```python
# Looking for missing data

for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
df = df.fillna(0)
```

```python
# Looking at different Data types

df.dtypes
```

```python
# Change some data types 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python

df.isna().sum().plot(kind="bar")
plt.show()
```

```python
#Drop Duplicates

df.drop_duplicates()
```

```python
#Scatter plot of Budget vs Gross

fig, ax = plt.subplots()
ax.scatter(df["budget"], df["gross"])
ax.set_title('Budget vs Gross Earnings')
ax.set_xlabel("Gross Earnings")
ax.set_ylabel("Budget for Film")
plt.show()
```

```python
df.head()
```

```python
# Plot Budget vs Gross with Seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color' : 'green'}, line_kws={'color' : 'red'} )
```

```python
#Correlation

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
#Company feature

df_num= df

for col in df_num.columns:
    if(df_num[col].dtype == 'object'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes
        
df_num
```

```python
correlation_matrix = df_num.corr()

sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
plt.show()
```

```python
df_num.corr()
```

```python
corr_mat = df_num.corr()
corr_pairs= corr_mat.unstack()
corr_pairs
```
